## Ridge Regularization


```python
from sklearn import linear_model  
for i in range (-3,4):
    alpha=10**i
    rm=linear_model.Ridge(alpha=alpha)
    ridge_model=rm.fit(xtrain,ytrain)
    ypred=ridge_model.predict(xtest)
    print('r^2 score =',Lreg.score(xtest,ytest))
    print('Rmse value=',np.sqrt(mse(ytest,ypred)))
        
    t1='Ridge Regression for alpha='
    t2=str(alpha)
    title = t1+t2
    sb.regplot(ytest,ypred)
    plt.xlabel('actual_revenue')
    plt.ylabel('predicted_revenue')
    plt.title(title)
    plt.show()   
    print('\n\n')
        
```

```python
testfinal['budget_sqrt'] = np.sqrt(testfinal.budget)
testfinal['budget_into_pop']=testfinal.budget*testfinal.popularity
testfinal['budget_collection']=testfinal.budget*testfinal.has_collection
testfinal['pop_collection']=testfinal.popularity*testfinal.has_collection
testfinal['budget_sq_pop']=np.square(testfinal.budget*testfinal.popularity)
testfinal['pop_sqr']=np.square(testfinal.popularity)
testfinal['_budget_rating_ratio'] = testfinal['budget']/testfinal['rating']
testfinal['budget_into_popluarity'] = testfinal['budget']*testfinal['popularity']
testfinal['_budget_popularity_ratio'] = testfinal['budget']/testfinal['popularity']
testfinal['_budget_into_rating'] = testfinal['budget']*testfinal['rating']
testfinal['_budget_into_popularity2'] = testfinal['budget']*testfinal['popularity2']
testfinal['_budget_into_totalVotes'] = testfinal['budget']*testfinal['totalVotes']
testfinal['_pop_into_totalVotes'] = testfinal['popularity']*testfinal['totalVotes']
testfinal['_pop_into_popularity2'] = testfinal['popularity']*testfinal['popularity2']
testfinal['_pop_into_rating'] = testfinal['popularity']*testfinal['rating']
testfinal['pop2_sqr']=np.square(testfinal.popularity2)
testfinal['rating_sqr']=np.square(testfinal.rating)
testfinal['totalsqrVotes']=np.square(testfinal.totalVotes)
testfinal['totalVotes_dev_rating']=testfinal.totalVotes/testfinal.rating
testfinal['pop2_into_rating']=testfinal.popularity2*testfinal.rating
testfinal['_pop2_into_totalVotes'] = testfinal['popularity2']*testfinal['totalVotes']
```

```python
from sklearn.model_selection import KFold
random_seed = 2045
k = 10
fold = list(KFold(k, shuffle = True, random_state = random_seed).split(X))
np.random.seed(random_seed)
```

```python
def xgb_model(trn_x, trn_y, val_x, val_y, test, verbose) :
    
    params = {'objective': 'reg:linear', 'eta': 0.01, 'max_depth': 3, 'subsample': 0.7, 'colsample_bytree': 0.8,  
              'eval_metric': 'rmse', 
              'seed': random_seed, 
              'silent': True,
    }
    
    record = dict()
    model = xgb.train(params
                      , xgb.DMatrix(trn_x, trn_y)
                      , 10000
                      , [(xgb.DMatrix(trn_x, trn_y), 'train'), (xgb.DMatrix(val_x, val_y), 'valid')]
                      , verbose_eval=verbose
                      , early_stopping_rounds=200
                      , callbacks = [xgb.callback.record_evaluation(record)])
    best_idx = np.argmin(np.array(record['valid']['rmse']))

    val_pred = model.predict(xgb.DMatrix(val_x), ntree_limit=model.best_ntree_limit)
    test_pred = model.predict(xgb.DMatrix(test), ntree_limit=model.best_ntree_limit)

    return {'val':val_pred, 'test':test_pred, 'error':record['valid']['rmse'][best_idx], 'importance':[i for k, i in model.get_score().items()]}
```

```python

result_dict = dict()
val_pred = np.zeros(X.shape[0])
test_pred = np.zeros(testfinal.shape[0])
final_err = 0
verbose = False

for i, (trn, val) in enumerate(fold) :
    print(i+1, "fold.    RMSE")
    
    trn_x = X.loc[trn, :]
    trn_y = Y[trn]
    val_x = X.loc[val, :]
    val_y = Y[val]
    
    fold_val_pred = []
    fold_test_pred = []
    fold_err = []
    
    result = xgb_model(trn_x, trn_y, val_x, val_y, testfinal, verbose)
    fold_val_pred.append(result['val'])
    fold_test_pred.append(result['test'])
    fold_err.append(result['error'])
    test_pred += np.mean(np.array(fold_test_pred),axis = 0)/k
    print("xgb model.", "{0:.5f}".format(result['error']))
```
